```python
# socket is the way two or more computers connect with each other
import socket
import select
import sys

#### chat application ####

HEADER_SIZE = 10
HOST = "10.0.0.200"
PORT = 9999

#https://medium.com/vaidikkapoor/understanding-non-blocking-i-o-with-python-part-1-ec31a2e2db9b
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM) # create socket
server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

server_socket.bind((HOST, PORT)) # bind socket
server_socket.listen() # listen for connections

# list of inputs to monitor (select.select)
inputs = [server_socket] # as clients connect, get stored in this list

clients = {} # identification: socket is the key and user data is going to be the value

# receive messages
def msg_receive(conn): # conn is the client socket
    try:
        msg_header = conn.recv(HEADER_SIZE) # give us the header
        
        if not len(msg_header): # if no data is received
            return False
        
        msg_length = int(msg_header.decode("utf-8").strip())
        return {"header": msg_header, "data": conn.recv(msg_length)}
        
    except:
        return False

        
# get header and data of user
while True:
    read_sockets, write_sockets, exception_sockets = select.select(inputs, [], inputs)
    for notified_socket in read_sockets:     # https://pymotw.com/2/select/
        if notified_socket == server_socket: # accept and handle connection when someone connects
            conn, address = server_socket.accept() # creates a new socket and returns it with client's address
            
            user = msg_receive(conn) # user variable has data and header of the client
            if user is False: # in case someone disconnects
                continue
            
            inputs.append(conn)
            # https://www.digitalocean.com/community/tutorials/understanding-dictionaries-in-python-3
            clients[conn] = user #save user data into dictionary
            
            # address[0] gives IP, address[1] gives port
            print(f"Accepted new connection from {address[0]}:{address[1]} username:{user['data'].decode('utf-8')}")
        
        else:
            message = msg_receive(notified_socket)
                  
            if message is False:
                print(f"Closed connection from {clients[notified_socket]['data'].decode('utf-8')}")
                inputs.remove(notified_socket)
                del clients[notified_socket]
                continue
        
            user = clients[notified_socket]
            print(f"Received message from {user['data'].decode('utf-8')}: {message['data'].decode('utf-8')}")
        
            for conn in clients:
                  if conn != notified_socket:
                      conn.send(user['header'] + user['data'] + message['header'] + message['data'])
    
    for notified_socket in exception_sockets:
        inputs.remove(notified_socket)
        del clients[notified_socket]
              
              
             
'''
# Create a socket
def create_socket():
    try:
        global host
        global port
        global s
        host = ""
        port = 9999
        # create a socket
        s = socket.socket()
    # if s gets an error    
    except socket.error as msg:
        print("Socket creation error" + str(msg))

# Binding the socket and listening for connections
def bind_socket():
    try:
        # whenever you want to access the variables that are global under the function
        # in python, you have to declare them again globally using "global"
        global host
        global port
        global s
        
        print("Binding the Port:  " + str(port))
        # tuple format
        s.bind((host, port))
        # 5 is the number connections that it will connect until it hits an error
        s.listen(5)
    
    except socket.error as msg:
        print("Socket creationg error" + str(msg) + "\n" + "Retrying...")
        bind_socket()

# Establish connection with a client (socket must be listening)
def socket_accept():
    # function for accepting the connection
    # conn = connection object
    # address (is a list, first element is the IP address)
    conn, address = s.accept()
    print("Connection has been established |" + "IP " + address[0] + " | Port" + str(address[1]))
    send_commands(conn)
    # closing the connection
    conn.close()
    
# Send commands to client/victim or a friend
def send_commands(conn):
    # infinite loop to send commands more than once
    while True: 
        cmd = input()
        if cmd == "quit":
            conn.close()
            s.close()
            sys.exit()
        # encode the cmd input into bytes to send as you can't send a string
        if len(str.encode(cmd)) > 0:
            # the part that actually sends the code to the other clients
            conn.send(str.encode(cmd))
            # whenever data is received, you have to convert it from byte -> str
            # 1024 is the byte chunks that we receive at a time.
            # conn.recv is used to receive information back from the client.
            # utf-8 is the coding type. utf-8 converts bytes into a format that can be
            # translated into str
            client_response = str(conn.recv(1024), "utf-8")
            print(client_response, end="")
            
def main():
    create_socket()
    bind_socket()
    socket_accept()
    
main()
'''
```
